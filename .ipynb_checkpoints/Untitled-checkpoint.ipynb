```python
import pandas as pd
import numpy as np

df = pd.read_csv('fwpHarvestEstimatesReport.csv')
```

```python
df
```

```python
def divide_two_cols(df_sub):
    hunters = float(df_sub['Hunters'].sum())
    if hunters <= 0:
        hunters += 1
    df_sub['Success Rate'] = hunters / float(df_sub['Total Harvest'].sum())
    return df_sub
df.groupby(['License Year']).apply(divide_two_cols)
```
